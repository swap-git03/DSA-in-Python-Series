```python
# 1. Electricity Bill(Slab)

# First 100 units ₹5/unit
# Next 100 units ₹7/unit

# Above 200 units ₹10/unit
# Calculate total bill.

units = float(input("Enter Units :"))
print(f"units are {units}")

if (units>200):
  amount = (units-200) * 10+ 100*7 + 100*5

elif (units>100):
  amount = (units-100) * 7 + 100*5

else:
  amount  = units* 5


print(f"total amount is {amount}")


```

```python
# 	4. E-Commerce Discount
# 	≥ 5000 → 20% discount
# 2000–4999 → 10% discount
# Premium member → Extra 5%

amount = float(input("enter the amount "))
premium = input("are you premium user? (yes/no): ").lower()

if (amount >= 5000):
  discount = 0.20

elif (amount >= 2000 and amount<=4999):
  discount = 0.10

else:
  discount = 0

# extra premium discount
if premium == "yes":
    discount += 0.05

final_amount = amount - (amount * discount)
print(f"the final amount is {final_amount}")




# premium = input("are you premium user ?")
```

```python
#7. BMI Checker	BMI = weight / (height × height)
# <18.5 → Underweight
# 18.5–24.9 → Normal
# 25–29.9 → Overweight
# ≥30 → Obese

weight = float(input("Enter weight :"))
height = float(input("Enter height :"))

bmi = weight / (height * height)
print(bmi)

if(bmi<18.5):
  print("underweight")
elif(bmi>=18.5 and bmi<25):
  print("Normal")
elif(bmi>=25 and bmi<30):
  print("Overweight")
else:
  print("Obese")
```
