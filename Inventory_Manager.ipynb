```python
def menu():
  print("\n------------------------")
  print("\nShop Inventory Manager")
  print("\n-----------------------")
  print("\n 1. Add New Product")
  print("\n 2. Sell Product")
  print("\n 3. Restock Product")
  print("\n 4. View STock Report")
  print("\n 5. Exit")
  print("\n------------------------")
  print("\n Enter Your CHoiceeee")



def add_product(product_name, product_price, product_quantity):
    print("\n Add Product")

    name = input("Enter product name: ")
    if name in product_name:
      print(f"{name} already exists in inventory!!!")
    else:
      price = float(input("Enter price per unit: "))
      quantity = int(input("Enter quantity: "))

      product_name.append(name)
      product_price.append(price)
      product_quantity.append(quantity)


def sell_product(product_name, product_price, product_quantity):
    print("\n Sell Product")
    name = input("Enter product name: ")
    if name not in product_name:
      print(f"{name} not found in inventory!!!")
    else:
      index = product_name.index(name)
      quantity_to_sell = int(input("Enter quantity to sell : "))
      if quantity_to_sell > product_quantity[index]:
        print(f"Not enough quantity of prduct is availble")
      else:
        product_quantity[index] -= quantity_to_sell
        sale_value = quantity_to_sell * product_price[index]

        print(f"we hve Sold {quantity_to_sell}units of {name}")
        print(f"sale value : {sale_value} rupee")
        print(f" Remaining stock is {product_quantity[index]}")


def restock_product(product_name, product_price, product_quantity):
  print("\n------Restock Product------")
  name = input("Enter product name: ")
  if name not in product_name:
    print(f"{name} not found in inventory!!!")
  else:
    index = product_name.index(name)
    quantity_to_add = int(input("Enter quantity to add: "))
    product_quantity[index] += quantity_to_add
    print(f"{name} restocked!!! New quantity {product_quantity[index]}")


def view_stock(product_name, product_price, product_quantity):
     print("\n------------------------")
     print("\nStock Reportttttttt")
     print("\n-----------------------")


     if len(product_name) ==0:
        print("No peroductsss in inventorty")
        return

     total_value = 0
     print("No.   Product   Price   Quantity    Status")
     print("\n-----------------------")


     for i in range(len(product_name)):
        quantity = product_quantity[i]

        if quantity == 0:
          status = "Out of Stock"
        elif quantity < 5:
          status = "Low Stock"
        else:
          status = "Available"

        print(f"{i+1}.   {product_name[i]}    {product_price[i]}   {quantity}    {status}")
        total_value += product_price[i]*quantity

     print("\n------------------------")
     print(f"Total Products: len({product_name})")
     print(f"Total Inventory Value: {total_value}")


product_name=[]
product_price=[]
product_quantity=[]


while True:
    menu()
    choice = int(input("Enter Your Choice: "))

    if choice == 1:
        add_product(product_name, product_price, product_quantity)

    elif choice == 2:
        sell_product(product_name, product_price, product_quantity)

    elif choice == 3:
        restock_product(product_name, product_price, product_quantity)

    elif choice == 4:
        view_stock(product_name, product_price, product_quantity)

    elif choice == 5:
        break



```
